```python
#Write a Python program to calculate the sum of all numbers from 1 to a number entered by the user.
def addition(n: int)->int:
    total = 0
    for i in range(1, n+1):
        total += i
    return total

try:    
    n = int(input("Enter a number: "))
    total = addition(n)
    print(f"Addition of 1 to {n} is: {total}")
except:
    print("Enter valid number")
```

```python
### **Question 2:**
#Write a Python program to check if a number is even or odd.**
def evenOrOdd(n: int):
    if n%2==0:
        print(f"{n} is even number")
    else:
        print(f"{n} is odd number")
    
try:    
    n = int(input("Enter a number: "))
    evenOrOdd(n)
except:
    print("Please enter valid number")
```
